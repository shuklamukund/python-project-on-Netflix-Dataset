### How many Movies got the 'TV-14' rating in canada?

```python
data[(data['type']=='Movie') & (data['rating']=='TV-14') & (data['country']=='Canada')].head(2)
```

(data['type']=='Movies') & (data['rating']=='TV-14') is ok but data['type']=='Movies' & data['rating']=='TV-14' is not ok becoz the & operator has higher precedence than the comparison operators (==), leading to an incorrect interpretation of the statement. 

```python
data[(data['type']=='Movie') & (data['rating']=='TV-14') & (data['country']=='Canada')].shape
```
